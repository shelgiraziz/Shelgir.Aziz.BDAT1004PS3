```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame with the relevant data
data = {
    'Year': [1900, 1950, 2000],
    'Marriages_per_1000': [9.3, 11.0, 8.2],
    'Divorces_per_1000': [0.7, 2.5, 4.0]
}
df = pd.DataFrame(data).set_index('Year')

# Create the vertical bar chart
ax = df.plot(kind='bar', figsize=(8, 6), width=0.8, color=['blue', 'red'])

# Set the title and axis labels
ax.set_title('Marriages and Divorces per Capita in the U.S.')
ax.set_xlabel('Year')
ax.set_ylabel('Number per 1000')

# Add the legend
ax.legend(['Marriages', 'Divorces'])

# Show the plot
plt.show()
```
