```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
emperors = pd.read_csv('roman-emperor-reigns.csv')

# Calculate the number of emperors who were assassinated
num_assassinated = (emperors['Cause_of_Death'] == 'Assassinated').sum()

# Calculate the number of emperors who died of natural causes or other reasons
num_not_assassinated = len(emperors) - num_assassinated

# Create a pie chart
labels = ['Assassinated', 'Not assassinated']
sizes = [num_assassinated, num_not_assassinated]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')

# Add a title
plt.title('Fraction of Roman Emperors who were Assassinated')

# Display the chart
plt.show()
```
